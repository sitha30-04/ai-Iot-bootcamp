```python
# step-3-- handling missing data and Nan
#scenario-1:missing values in numeric column
#solution:calculate the avg and mean values of the data
#scenario-2:missing values in non numerical columns
#solution:fil the empty repeated values
#scenario-3:if you having missing values in huge dataset
import numpy as np
from sklearn.impute import SimpleImputer # it is a class which was imported from sklearn.impute
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x[:,1:3])
x[:,1:3]=imputer.transform(x[:,1:3])
```

```python
#step 4--Handling the catagorial data
## converting catagorical data into numeric data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough' )
np.remainder='passthrough'
x=ct.fit_transform(x)
```
